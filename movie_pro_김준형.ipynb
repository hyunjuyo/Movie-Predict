## Weight Rating
```
WR = (v/(v+m)*R) + (m/(m+v)*C)
```

````python
df_inner_id_v2.info()
````

````python
C = df_inner_id_v2["vote_average"].mean()
C
````

````python
m =df_inner_id_v2["vote_count"].quantile(0.95)
m
````

````python
def weighted_rating(x):
    v = x["vote_count"]
    R = x["vote_average"]
    return (v/(v+m)*R) + (m/(m+v)*C)
````

````python
df_inner_id_v2["WR"] = df_inner_id_v2.apply(weighted_rating,axis=1)
````

````python
df_inner_id_v2.drop(columns=["vote_average","vote_count"],inplace=True)
````
